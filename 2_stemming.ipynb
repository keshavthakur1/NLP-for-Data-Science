# Stemming


Stremming is the process of reducing a word to its stem that affixes to suffix  and preffix or to the roots of words known as a lemma. Stemming is a important in natural language understanding(NLU) and natural language processing(Nlp)

```python
#  classification problem
#  comment of product is a positive 
# review -----> eating , eat , eaten [goging, gone, goes ]
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

# PorterStremmer

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
for word in words:
    print(word+"---->"+stemmer.stem(word))   # what disadvantage of stemming?   # it is not giving proper root word 
```

```python
stemmer.stem('congratulations')  # it is giving congratul which is not root word  so this major major disadvantage of stemming
```

```python
stemmer.stem('running')  # run
```

### RegexpStemmer class


NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
for word in words:
    print(word+"---->"+reg_stemmer.stem(word))
```

```python
reg_stemmer.stem("fairly"),reg_stemmer.stem("sportingly") # why i got fair and sport  because min=4  so it will not stem if length is less than 4
```

### Snowball Stemmer

It is a stemming algorithm which is also known as the Porter2 stemming algorithm as it is a better version of the Porter Stemmer since some issues of it were fixed in this stemmer.

```python
from nltk.stem import SnowballStemmer
```

```python
snowballsstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowballsstemmer.stem(word))
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()     

print(stemming.stem("fairly"))
print(stemming.stem("sportingly"))
```

```python
snowballsstemmer.stem("fairly"),snowballsstemmer.stem("sportingly")
```

```python
snowballsstemmer.stem('goes')
```
