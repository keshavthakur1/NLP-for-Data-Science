```python
corpus ="Hello welcome, to keshav thakur Nlp Tutorial. PLease do watch the entire video! to become more familar with nlp. Also do like and subscribe. Thank you"
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
print(documents)   # it is converted into list of sentences
```

```python
import nltk

# For older versions:
# nltk.download('punkt')

# For newer versions (Python 3.12 + NLTK 3.8+):
nltk.download('punkt_tab')  # what is punkt_tab?    punkt_tab is a version of the Punkt tokenizer that is optimized for tab-separated text.

from nltk.tokenize import sent_tokenize

documents = sent_tokenize(corpus)
print(documents)
type(documents)
```

```python
for senetence in documents:
    print(senetence)
```

```python
from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
print(words)
print(len(words))
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(corpus))  # there is is little bit difference in output if u observe carefully  in tokenization fullstop and not treated as a single token but at end of sentence it is treated as a single token.


```
